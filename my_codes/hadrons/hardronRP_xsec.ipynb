```python
def mean(x,y):
    numerator = 0
    denominator = sum(y)
    for i in range(0,len(x)):
        numerator += x[i]*y[i]
        mean = numerator/denominator
    return mean

def std(x,y):
    num = 0
    denominator = sum(y)
    for i in range(0,len(x)):
        num += (x[i]-mean(x,y))**2
        std = np.sqrt(num/(denominator-1))
    return std

print("JER = ", mean(xnew,y))
print("Standard Deviation = ", std(xnew,y))
```
