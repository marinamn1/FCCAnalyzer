```python
import uproot
import matplotlib.pyplot as plt
import numpy as np
import hist
from hist import Hist
import mplhep as hep
import math
import plots

f = uproot.open("/home/submit/marinamn/fccee/FCCAnalyzer/tmp/output_hadron_xsec.root")
```

```python
sample_list = ["kkmc_ee_uu_ecm91p2_noBES_noISR","wz3p8_ee_uu_ecm91p2_noBES_noISR", "kkmc_ee_uu_ecm91p2", "wz3p8_ee_uu_ecm91p2", "wzp6_ee_uu_ecm91p2"]
label_list = [r"$e^{+} e^{-} \rightarrow u \bar{u}$, KKMC Pythia 8, no BES/ISR/FSR", r"$e^{+} e^{-} \rightarrow u \bar{u}$, Whizard Pythia 8, no BES/ISR/FSR", r"$e^{+} e^{-} \rightarrow u \bar{u}$, KKMC Pythia 8", r"$e^{+} e^{-} \rightarrow u \bar{u}$, Whizard Pythia 8", r"$e^{+} e^{-} \rightarrow u \bar{u}$, Whizard Pythia 6"]
color_list = ["red", "blue", "lightskyblue", "violet", "darkgreen"]

for i in range(0,5):
    discard2_no = f[f"{sample_list[i]}/discard2RP_no_vis"].to_hist()
    discard1RP_e = f[f"{sample_list[i]}/discard1RP_e_vis"].to_hist()
    discard1_e = f[f"{sample_list[i]}/discard1norm_energy"].to_hist()
    discard1_theta = f[f"{sample_list[i]}/discard1RP_theta"].to_hist()
    discard1_cost = f[f"{sample_list[i]}/discard1cos_thetat"].to_hist()
    discard1_thrust = f[f"{sample_list[i]}/discard1thrust"].to_hist()
    
    
    hep.histplot(discard2_no, label = label_list[i], color = color_list[i], histtype = "fill")
    plt.xlim(0,60)
    plt.xlabel(r"Number of Reconstructed Particles")
    plt.legend(loc = "lower center")
    plt.yscale("log")
    plt.title(r"$\sqrt{s} = 91.2 $ GeV, 75 ab$^{-1}$", loc = "right")
    plt.title("FCC-ee simulation", loc = "left", weight = "bold")
    #plt.text(48,1.8e8,r"Events with $E_{vis}$/$ \sqrt{s}$ < 0.4", verticalalignment='top', horizontalalignment='right')
    plt.ylabel("Events")
    plt.savefig(f"images/discard2_no_{sample_list[i]}.png")
    plt.close()
    
    hep.histplot(discard1RP_e, label = label_list[i], color = color_list[i], histtype = "fill")
    plt.xlabel(r"$ E_{particle} $ / $\sqrt{s}$")
    plt.xlim(0, 0.55)
    plt.legend(loc = "lower center")
    plt.yscale("log")
    plt.title(r"$\sqrt{s} = 91.2 $ GeV, 75 ab$^{-1}$", loc = "right")
    plt.title("FCC-ee simulation", loc = "left", weight = "bold")
    #plt.text(0.54,1e9,"Events with Number of Reconstructed Particles < 14", verticalalignment='top', horizontalalignment='right')
    plt.ylabel("Particles")
    plt.savefig(f"images/discard1RP_e_{sample_list[i]}.png")
    plt.close()
    
    hep.histplot(discard1_e, label = label_list[i], color = color_list[i], histtype = "fill")
    plt.xlabel(r"$ E_{vis} $ / $\sqrt{s}$")
    plt.xlim(-0.01, 1.03)
    plt.legend(loc = "upper center")
    plt.yscale("log")
    plt.title(r"$\sqrt{s} = 91.2 $ GeV, 75 ab$^{-1}$", loc = "right")
    plt.title("FCC-ee simulation", loc = "left", weight = "bold")
    #plt.text(1,1.8e8,"Events with Number of Reconstructed Particles < 14", verticalalignment='top', horizontalalignment='right')
    plt.ylabel("Events")
    plt.savefig(f"images/discard1_e_{sample_list[i]}.png")
    plt.close()
    
    hep.histplot(discard1_theta, label = label_list[i], color = color_list[i], histtype = "fill")
    plt.xlim(0,3.15)
    plt.yscale("log")
    plt.xlabel(r"$ \theta_{particle} $ [radians]")
    plt.legend(loc = "lower center")
    plt.title(r"$\sqrt{s} = 91.2 $ GeV, 75 ab$^{-1}$", loc = "right")
    plt.title("FCC-ee simulation", loc = "left", weight = "bold")
    #plt.text(2.7,1e8,"Events with Number of Reconstructed Particles < 14", verticalalignment='top', horizontalalignment='right')
    plt.ylabel("Particles")
    plt.savefig(f"images/discard1RP_theta_{sample_list[i]}.png")
    plt.close()
    
    hep.histplot(discard1_cost, label = label_list[i], color = color_list[i], histtype = "fill")
    plt.xlabel(r"| cos ($ \theta_{t} $) |")
    plt.xlim(-0.05,1.05)
    plt.yscale("log")
    plt.legend(loc = "upper center")
    plt.title(r"$\sqrt{s} = 91.2 $ GeV, 75 ab$^{-1}$", loc = "right")
    plt.title("FCC-ee simulation", loc = "left", weight = "bold")
    #plt.text(0.8,1.6e8,"Events with Number of Reconstructed Particles < 14", verticalalignment='top', horizontalalignment='right')
    plt.ylabel("Events")
    plt.savefig(f"images/discard1_cost_{sample_list[i]}.png")
    plt.close()

    hep.histplot(discard1_thrust, label = label_list[i], color = color_list[i], histtype = "fill", flow = None)
    plt.xlabel("| Thrust |")
    plt.legend(loc = "upper center")
    plt.yscale("log")
    plt.xlim(0.6,1.01)
    plt.title(r"$\sqrt{s} = 91.2 $ GeV, 75 ab$^{-1}$", loc = "right")
    plt.title("FCC-ee simulation", loc = "left", weight = "bold")
    #plt.text(0.95,2e8,"Events with Number of Reconstructed Particles < 14", verticalalignment='top', horizontalalignment='right')
    plt.ylabel("Events")
    plt.savefig(f"images/discard1_thrust_{sample_list[i]}.png")
    plt.close()
    
```
