```python
import uproot
import matplotlib.pyplot as plt
import numpy as np
import hist
from hist import Hist
import mplhep as hep

f = uproot.open("/home/submit/marinamn/fccee/FCCAnalyzer/tmp/output_xsec_example.root")
f.keys()
```

```python
h = f['p8_ee_Zmumu_ecm91/muon_leading;1'].to_hist()
htau = f['p8_ee_Ztautau_ecm91/muon_leading;1'].to_hist()
hb = f['wzp6_gaga_mumu_5_ecm91p2/muon_leading;1'].to_hist()
```

```python

y,x = h.to_numpy()

ytau,xtau = htau.to_numpy()
yb,xb = hb.to_numpy()

xnew = x/45.5
xtaunew = xtau/45.5
xbnew = xb/45.5

plt.stairs(y+ytau, xnew, label = r"$e^{+} e^{-} \rightarrow \mu^{+} \mu^{-}$", fill = True)
plt.stairs(ytau, xtaunew, label = r"$e^{+} e^{-} \rightarrow \tau^{+} \tau^{-}$", fill = True)
plt.stairs(yb, xbnew, label = r"$e^{+} e^{-} \rightarrow e^{+} e^{-} \mu^{+} \mu^{-}$", fill = True)
plt.legend(loc = "upper left")
plt.yscale("log")
plt.xlim(0,1.06)
plt.xlabel(r"$P_{\mu}/E_{b}$")
plt.ylabel("Events")
plt.title(r"$\sqrt{s} = 91 GeV, 150 ab^{-1}$", loc = "right")
plt.title("FCC-ee simulation", loc = "left", weight = "bold", style = "italic")
#plt.savefig("images/papercuts_p.png")
```

```python
cuts = f['p8_ee_Zmumu_ecm91/cutFlow;1'].to_hist()
cutstau = f['p8_ee_Ztautau_ecm91/cutFlow;1'].to_hist()
cutsb = f['wzp6_gaga_mumu_5_ecm91p2/cutFlow;1'].to_hist()
hep.histplot([cutstau,cuts, cutsb])
plt.xlim(0,8)
plt.yscale("log")
```

```python
acol = f['p8_ee_Zmumu_ecm91/acol'].to_hist()
acoltau = f['p8_ee_Ztautau_ecm91/acol'].to_hist()
acolb = f['wzp6_gaga_mumu_5_ecm91p2/acol;1'].to_hist()

hep.histplot([acolb,acoltau,acol], stack = True, histtype = "fill", 
             label = [r"$e^{+} e^{-} \rightarrow e^{+} e^{-} \mu^{+} \mu^{-}$",
                      r"$e^{+} e^{-} \rightarrow \tau^{+} \tau^{-}$", 
                      r"$e^{+} e^{-} \rightarrow \mu^{+} \mu^{-}$"])
plt.xlim(0,1.6)
plt.yscale("log")
plt.legend(loc = "upper right")
plt.xlabel(r"Acollinearity ($\xi$) [rad]")
plt.ylabel("Events")
plt.title(r"$\sqrt{s} = 91 GeV, 150 ab^{-1}$", loc = "right")
plt.title("FCC-ee simulation", loc = "left", weight = "bold", style = "italic")
#plt.savefig("images/papercuts_acol.png")
```

```python
t = f['p8_ee_Zmumu_ecm91/cos_theta;1'].to_hist()
yt,theta = t.to_numpy()

ttau = f['p8_ee_Ztautau_ecm91/cos_theta;1'].to_hist()
tb = f['wzp6_gaga_mumu_5_ecm91p2/cos_theta;1'].to_hist()
yttau,thetatau = ttau.to_numpy()
ytb,xtb=tb.to_numpy()

plt.stairs(yt+ytb+yttau, theta, label = r"$e^{+} e^{-} \rightarrow \mu^{+} \mu^{-}$", fill = True)
plt.stairs(yttau+ytb, thetatau, label = r"$e^{+} e^{-} \rightarrow \tau^{+} \tau^{-}$", fill = True)
plt.stairs(ytb, xtb, label = r"$e^{+} e^{-} \rightarrow e^{+} e^{-} \mu^{+} \mu^{-}$", fill = True)
plt.xlim(-1.1,1.1)
plt.yscale("log")
plt.legend()
plt.xlabel(r"cos(${\theta}$)")
plt.ylabel("Events")
plt.title(r"$\sqrt{s} = 91 GeV, 150 ab^{-1}$", loc = "right")
plt.title("FCC-ee simulation", loc = "left", weight = "bold", style = "italic")
#plt.savefig("images/papercuts_cos_theta.png")
```
