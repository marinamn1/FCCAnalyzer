```python
def acceptance_error(n_sel, n_tot):
    return np.sqrt(n_sel*(1-(n_sel/n_tot)))/n_tot
def cross_section_error(n_sel, n_tot, l):
    a = n_sel/n_tot
    return {np.sqrt((n_sel/(a**2*l**2))+(n_sel**2*acceptance_error(n_sel, n_tot)**2/(a**4*l**2)))}
def acceptance(n_sel, n_tot):
    return f"Acceptance: {n_sel/n_tot} ± {acceptance_error(n_sel, n_tot)}"
def cross_section(n_sel, n_tot, l):
    a = n_sel/n_tot
    result = n_sel/(a*l)
    return f"Cross Section: {result} ± {cross_section_error(n_sel, n_tot, l)}"
```

```python
xsec = np.array([10000,17400,34500,27500,15300])
xdata = np.array([89.5,90.2,91.2,92,93])
yer1 = [0.023,0.03,0.021,0.038,0.029]
yer2 = [0.018,0.024,0.034,0.03,0.027]
yer3 = [0.028,0.037,0.019,0.047,0.035]

def breit_wigner(x, mass, width, k):
    s = x**2
    xsec = k*s*(width**2)/(((mass*width)**2)*(((s-(mass**2))**2)+((s*width/mass)**2)))
    return xsec


popt, pcov = scipy.optimize.curve_fit(breit_wigner, xdata, xsec, p0 = (91,2.5,215))
x = np.linspace(88, 94, 1000)

ratio = xsec/breit_wigner(xdata, *popt)
fig, ax = plt.subplots(2, sharex=True, gridspec_kw={'height_ratios': [3, 1]})
ax[0].scatter(xdata,xsec, label = "MC Data")
ax[0].set_ylabel(r"$\sigma $ [pb]")
ax[1].errorbar(xdata,ratio,yerr = yer1/breit_wigner(xdata, *popt), fmt = "o")
ax[0].plot(x, breit_wigner(x, *popt), color = "paleturquoise", label = "Breit-Wigner fit")
ax[0].legend()
ax[1].scatter(xdata,ratio)
ax[1].set_ylim(0.9,1.15)
ax[1].axhline(1.0,linestyle='dashed',color='gray')
plt.xlabel(r"$\sqrt{s} $ [GeV]")
ax[1].set_ylabel("ratio")
fig.tight_layout()
#plt.savefig("images/hardrons_xsec.png")

print(popt)
print(np.sqrt(pcov[0][0]),np.sqrt(pcov[1][1]))
```
